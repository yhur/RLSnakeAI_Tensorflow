```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Input
import os
import json
import numpy as np
```

```python
input_size = 14
output_size = 3
org_model = Sequential([
    Input(shape=(input_size, )),
    Dense(256, activation='relu'),
    Dense(output_size, activation='linear')
])
new_model = Sequential([
    Input(shape=(input_size + 3, )),
    Dense(256, activation='relu'),
    Dense(output_size, activation='linear')
])
```

```python
def load(model, dir_name='model'):
    model_file = os.path.join(dir_name, 'model.keras')
    info_file  = os.path.join(dir_name, 'model.json')
    for f in (model_file, info_file):
        if not os.path.exists(f):
            print(f"Warning: Missing file: {f}")
            return

    model.set_weights(load_model(model_file).get_weights())
    
    with open(info_file, 'r') as f:
        info_json = json.load(f)
        record = info_json['record'] if 'record' in info_json else 0
        n_games = info_json['n_games'] if 'n_games' in info_json else 0
   
    return record, n_games
```

```python
def save(model, score, n_games, dir_name='model'):
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
    model_file = os.path.join(dir_name, 'model.keras')
    info_file  = os.path.join(dir_name, 'model.json')

    model.save(model_file)

    info_json = {'record': score, 'n_games': n_games}
    with open(info_file, 'w') as f:
        f.write(json.dumps(info_json))
```

```python
r, n = load(org_model)
```

```python
w=org_model.get_weights()
w[0] = np.vstack([w[0], np.zeros((3, 256))])
```

```python
new_model.set_weights(w)
```

```python
save(new_model, r, n, dir_name='newmodel')
```
